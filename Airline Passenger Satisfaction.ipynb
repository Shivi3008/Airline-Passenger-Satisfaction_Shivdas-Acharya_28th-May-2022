## Finding of people with higher class has more satisfaction

```python
train_data.head()
```

```python
class_satisfaction_df = train_data[['Class','satisfaction']].value_counts().sort_index()
```

```python
class_satisfaction_df
```

### - Percentage of people from Business Class that were satisfied:

```python
class_satisfaction_df['Business']['satisfied']*100/(class_satisfaction_df['Business']['satisfied']+class_satisfaction_df['Business']['neutral or dissatisfied'])
```

### - Percentage of people from Economy class that were satisfied:

```python
class_satisfaction_df['Eco']['satisfied']*100/(class_satisfaction_df['Eco']['satisfied']+class_satisfaction_df['Eco']['neutral or dissatisfied'])
```

### - Percentage of people from Economy Plus class that were satisfied:

```python
class_satisfaction_df['Eco Plus']['satisfied']*100/(class_satisfaction_df['Eco Plus']['satisfied']+class_satisfaction_df['Eco Plus']['neutral or dissatisfied'])
```

## From above results as per the class, Business class people are more satisfied than others
